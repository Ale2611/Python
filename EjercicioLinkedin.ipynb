```python
#Para prepara nuestra tabla de contingencia
np.random.seed(0)
df = df[df["Origin"].isin(["LAS","LAX"])]
df = df.sample(frac = 1)
df = df[1:10000]
```

```python
#Eso es una tabla de contingencia
df["BigDelay"] = df["ArrDelay"] > 30
observados = pd.crosstab(index=df["BigDelay"], columns=df["Origin"], margins=True)
observados
```

```python
from scipy.stats import chi2_contingency #Importación del paquete que va a realizar el test
```

```python
test = chi2_contingency(observados)
test
```

```python
esperados = pd.DataFrame(test[3])
esperados
```

```python
esperados_rel = round(esperados.apply(lambda r: r/len(df) *100,axis=1),2)
esperados_rel
```

```python
observados_rel = round(observados.apply(lambda r: r/len(df) *100,axis=1),2)
observados_rel

#no entendi nada. Tendría que ver la teoria del X cuadrado
```

```python
test[1] #si este número es inferior a 0,05, significa que hay relación entre variables
#si este número es superior a 0,05, signficia que no hay relaciones entre variables
```
