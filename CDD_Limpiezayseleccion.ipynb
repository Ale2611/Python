```python
#LIMPIEZA DE LOS DATOS
# Obtener los nombres de las columnas
nombres_columnas = dataframe.columns

# Crear una lista con los nuevos nombres de las columnas
nuevos_nombres = ['N.º', 'Usuario', 'Propietario', 'Seguidores_en_millones', 'Ocupación', 'País', 'Cuenta_de_marca']

# Asignar los nuevos nombres de las columnas al DataFrame
dataframe.columns = nuevos_nombres

# Verificar los nombres de las columnas para confirmar el cambio
print(dataframe.columns)
```

```python
#LIMPIEZA DE LOS DATOS
#Cambio los valores de la columna "Cuenta_de_marca"
dataframe['Cuenta_de_marca'] = dataframe['Cuenta_de_marca'].replace({'Sí': 1, 'No': 0})
```

```python
def mas_seguidores():
# Encontrar la celebridad con más seguidores
  indice_max_seguidores = dataframe['Seguidores_en_millones'].idxmax() #idmax me sirve para sacar el maximo de la columna seguidores que se encuentra en mi dataframe.
  celebridad_mas_seguidores = dataframe.loc[indice_max_seguidores] #loc me permite ver toda la linea
#Primera posibilidad de hacer
  print(celebridad_mas_seguidores['Propietario'])
  print(celebridad_mas_seguidores['Seguidores_en_millones'])
  print(celebridad_mas_seguidores['País'])

#Segunda posibilidad de hacer
# Seleccionar solo el nombre, la cantidad de seguidores y el país de origen
  nombre = celebridad_mas_seguidores['Propietario']
  seguidores = celebridad_mas_seguidores['Seguidores_en_millones']
  pais = celebridad_mas_seguidores['País']

# Imprimir los detalles seleccionados
  print("Nombre:", nombre)
  print("Cantidad de seguidores:", seguidores, "millones")
  print("País de origen:", pais)

```
