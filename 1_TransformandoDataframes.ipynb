```python
from sklearn.datasets import load_iris
import pandas as pd

# Cargar el conjunto de datos Iris
iris = load_iris()

# Cargar el conjunto de datos Iris
iris = load_iris()

# Crear un DataFrame de pandas
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```
